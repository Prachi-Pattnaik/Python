```python
import pandas as pd
df = pd.read_csv(r'egrocery.csv')
df.head()
```

```python
df.groupby(['Independent Variable']).mean()
```

```python
ct = pd.crosstab(df['Independent Variable'], df['Dependent Variable'])
ct
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

prob = 0.95
stat, p, dof, expected = chi2_contingency(ct)
critical = chi2.ppf(prob, dof)

print('1 - alpha:', prob)
print('D.o.F:', dof)
print('observed value:', stat)
print('critical value:', critical)
print('p value:', p)

print('')
print('RESULT:')

if abs(stat) >= critical:
    print('Reject H0: Depended')
else:
    print('Fail to reject H0: Independent')
```
